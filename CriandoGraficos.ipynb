```python
import pandas as pd

df = pd.read_csv('/content/imigrantes_canada.csv')
```

```python
df
```

```python
df.set_index('País',inplace=True)
```

```python
anos = list(map(str, range(1980,2014)))
anos
```

```python
brasil = df.loc['Brasil', anos]
brasil
```

```python
brasil.index
```

```python
brasil_dict = {'ano':brasil.index.tolist(),'imigrantes':brasil.values.tolist()}
dados_brasil = pd.DataFrame(brasil_dict)
```

```python
dados_brasil
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,4))
plt.plot(
    dados_brasil['ano'],
    dados_brasil['imigrantes']
    )
plt.title('Imigração do Brasil para o Canadá',loc='center')
plt.xlabel('Ano')
plt.ylabel('N° de imigrantes')
plt.xticks(list(map(str, range(1980,2011,5))))
plt.yticks([500,1000,1500,2000,2500,3000])
plt.show()
```

```python
df_comparacao =  df.loc[['Brasil','Argentina'], anos]
```

```python
df_comparacao
```

```python
df_comparacao = df_comparacao.T
```

```python
plt.plot(df_comparacao['Brasil'],label = 'Brasil')
plt.plot(df_comparacao['Argentina'],label ='Argentina')
plt.title('Imigração do Brasil e Argentina para o Canadá')
plt.xlabel('Ano')
plt.ylabel('Número de imigrantes')
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
plt.legend()
plt.show()
```

```python
cores = []
for pais in america_sul_sorted.index:
  if pais == 'Brasil':
    cores.append('green')
  else:
    cores.append('gray')


fig, ax = plt.subplots(figsize=(12,5))
ax.barh(america_sul_sorted.index, america_sul_sorted['Total'], color=cores)
ax.set_title('Imigração da America do Sul para o Canadá')
ax.set_xlabel('')
ax.set_ylabel('N° Imigrantes')
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)


for i,v in enumerate(america_sul_sorted['Total']):
  ax.text(v+20, i, str(v), color='Black',fontsize=10,ha='left',va='center')

ax.set_frame_on(False)
ax.get_xaxis().set_visible(False)
ax.tick_params(axis='both', which='both', length=0)

fig.savefig('imigração_america_sul.png', transparent= False, dpi=300, bbox_inches='tight')
```

```python
import seaborn as sns
```

```python
df
```

```python
sns.set_theme(palette='Set2')
```

```python
top_10 = df.sort_values('Total', ascending=False).head(10)
top_10
```

```python
sns.barplot(top_10, y=top_10.index,x='Total', orient='h',hue=top_10.index)
```

```python
ax = sns.barplot(top_10, y=top_10.index,x='Total', orient='h',hue=top_10.index)

ax.set(
    title='Top 10 Imigrantes Canadá\n1980-2013',
    xlabel='N° Imigrantes'
    )
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,4))

ax = sns.barplot(top_10, y=top_10.index,x='Total', orient='h',hue=top_10.index)
ax.set_title('Paises com maior imigração para o Canadá\n1980-2013', loc='left',fontsize=16)
ax.set_xlabel('N° de Imigrantes',fontsize=12)
ax.set_ylabel('')
plt.show()
```

```python
def gerar_grafico_paleta(palette):
  fig, ax = plt.subplots(figsize=(8,4))
  ax = sns.barplot(data=top_10, y=top_10.index, x='Total',orient='h',hue=top_10.index,palette=palette)
  ax.set_title('Paises com maior imigração para o Canadá')
  ax.set_xlabel('N° de imigrantes')
  ax.set_ylabel('')
  plt.show()
```

```python
gerar_grafico_paleta('flare_r')
```

```python
gerar_grafico_paleta('tab10')
```

```python
sns.set_theme(style='dark')
gerar_grafico_paleta('tab10')
```

```python
sns.set_theme(style='whitegrid')
gerar_grafico_paleta('tab10')
```
